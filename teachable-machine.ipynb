```python
from keras.models import load_model  
import cv2  
import numpy as np
```

Ahora vamos a utilizar la red neuronal entrenada (a través de Keras) para clasificar las imágenes capturadas en tiempo real que llegan desde la cámara web

```python
# Disable scientific notation for clarity
np.set_printoptions(suppress=True)
# Load the model and labels
model = load_model("keras_Model.h5", compile=False)
class_names = open("labels.txt", "r").readlines()

# Start the webcam
camera = cv2.VideoCapture(0)

while True:
    ret, image = camera.read()
    if not ret:
        break
    
    # Resize and process the image
    image = cv2.resize(image, (224, 224), interpolation=cv2.INTER_AREA)
    processed_image = np.asarray(image, dtype=np.float32).reshape(1, 224, 224, 3)
    processed_image = (processed_image / 127.5) - 1
    
    # Make prediction
    prediction = model.predict(processed_image)
    index = np.argmax(prediction)
    class_name = class_names[index].strip()
    confidence_score = prediction[0][index]

    # Print prediction and confidence score
    print("Class:", class_name[2:], end="")
    print("Confidence Score:", str(np.round(confidence_score * 100))[:-2], "%")

    # Create text to display
    text = f"{class_name}: {np.round(confidence_score * 100, 2)}%"
    
    # Put the text on the image
    cv2.putText(image, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)
    
    # Display the image
    cv2.imshow("Webcam Image", image)

    if cv2.waitKey(1) == 27:  # ASCII for the esc key
        break
        
camera.release()
cv2.destroyAllWindows()
```
